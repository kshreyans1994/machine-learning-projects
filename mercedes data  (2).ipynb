```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
md[md.select_dtypes(include="object").columns] =md[md.select_dtypes(include="object").columns].apply(le.fit_transform)
```

```python
md_x= md.iloc[:,1:378]
md_y=md.iloc[:,0]
```

```python
md_x
```

```python
md_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
md_x_train, md_x_test, md_y_train, md_y_test= train_test_split(md_x , md_y, test_size= .20)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression(normalize=True)
```

```python
reg.fit(md_x_train,md_y_train)
```

```python
pred_reg= reg.predict(md_x_test)
```

```python
error_reg= md_y_test-pred_reg
```

```python

Rsquare= reg.score(md_x_train,md_y_train)
Rsquare
```

```python

N=md_x_train.shape[0]
K=md_x_train.shape[1]

Adjusted_Rsquare= 1-(1-Rsquare)*(N-1)/(N-K-1)
print("value of Adjusted_Rsquare is ", Adjusted_Rsquare)
```

```python
MAPE=np.mean(np.abs(error_reg/md_y_test)*100)
MAPE
```

```python
Accuracy  = 100 - MAPE
print("Accuracy of the model is",np.round(Accuracy),"%")
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(md_x,md_y)
```

```python

print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_) 
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(md_x,md_y)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
prediction_lasso=lasso_regressor.predict(md_x_test)
prediction_ridge=ridge_regressor.predict(md_x_test)
```
